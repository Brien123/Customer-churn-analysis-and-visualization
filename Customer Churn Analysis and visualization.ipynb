# Exploratory Data Analysis 

```python
#checking how many customers churned
df.Churn.value_counts()  
```

```python
#plot of the customers churn ratio
sns.countplot(data=df, x='Churn') 
plt.show()
```

```python
#checking number of males and female customers 
sns.countplot(data=df, x='gender' ) 
plt.show()
```

```python
#checking ratio of senior citizens to non senior citizens
plt.figure(figsize=(15,9)) 
df.SeniorCitizen.value_counts().plot(kind='pie', shadow=True, startangle=90, autopct='%1.1f%%', explode=[0,0.1], labels=leg)
leg=['non senior citizens','Senior Citizen' ]
plt.legend(leg, loc='upper left')
plt.title('ratio of senior citizens to non senior citizens',loc='right')
plt.show()
```

```python
#Which citizens churned most 
sns.countplot(hue='SeniorCitizen', x='Churn', data=df)
plt.title('Churn Distribution') 
plt.legend(leg, loc='upper right')  
plt.show()
```

```python
#Which service lost the most customers  
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title('Which service lost the most customers') 
plt.show()
```

```python
#Most used payment methods
sns.countplot(x='PaymentMethod', data=df, hue='Churn')
plt.title('Most used payment methods')
plt.xticks(rotation=45)
plt.show()
```

```python
#Most used contracts
sns.countplot(x='Contract', data=df, hue='Churn')
plt.title('Most used Contract sizes')
plt.show()
```

```python
#What percentage of customers used phone services 
sns.countplot(x='PhoneService', data=df, hue='Churn')
plt.title('Distribution of Phone services')
plt.show()
```
